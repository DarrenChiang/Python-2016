# My first Scientific app in Python with Numpy
In this first example we will create a sine function and plot it. The equation is $y=3sin(2\pi x)$ where $x \in [-1, 1]$.

(1) Let's import the Numpy library.

```python
import numpy as np

# -- Create the values for x. This is a vector (a.k.a. list)
x = np.linspace(-1, 1, 1000)
y = 3 * np.sin(2 * np.pi * x)
```

(2) Import the graphic library and show the function y.

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# this is needed to show graphs inside the notebook

plt.plot(x, y)
plt.show()
```

(3) Calculate the differential of $y$ and show it.

```python
dy_dx = np.diff(y)

plt.plot(x[ : -1], dy_dx)

plt.xlabel('time (s)', fontsize = 14)
plt.ylabel('voltage (mV)', fontsize = 14)
plt.title('$y=2\pi cos(2\pi x)$', fontsize = 14)
plt.grid(True)
plt.savefig('test.png')
plt.show()
```

# Fun with statistics: app No 2

Let's create a random sample of 1000 points from a Poisson distribution.

```python
p = np.random.poisson(5, 1000)

plt.plot(p)
plt.show()
```

```python
total = 0
for n in p:
    total += n
print(total / len(p))
```

```python
print('The expected value (mean) of the Poisson distribution is ' + str(np.mean(p, axis = 0)))
```

```python
b, t, d = plt.hist(p, 15, normed = True)
```

```python
cdf = np.cumsum(b)
plt.plot(cdf)
plt.show()
```

```python
n = np.random.normal(loc = 86, scale = 10, size = 1000)
x, y, z = plt.hist(n, 120, normed = True)
plt.show()
```

```python
plt.plot(np.cumsum(x))
plt.show()
```
